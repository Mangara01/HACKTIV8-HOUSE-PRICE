#### iii. Penyelesaian Penjabaran Masalah

1. Berapa banyak rumah yang dijual di masing-masing kota di USA?

```python
# Proses pengambilan data rumah yang dijual di USA berdasarkan masing-masing kotanya
data1 = data_milestone[
    data_milestone[
        "country"] == "USA"].groupby([
    "country","city"])[
    "price"].count().sort_values(ascending=False)
```

```python
# Visualisasi data pada kota di USA dengan jumlah item penjualan rumah tertinggi hingga terendah
data1.plot(
    kind="bar", figsize=(10, 5), title="Jumlah rumah yang dijual di masing-masing kota di USA")
```

2. Kota mana dengan harga penjualan rumah paling besar?

```python
# Proses pengambilan data harga terbesar untuk setiap rumah pada masing-masing kota di USA
data2 = data_milestone[
    data_milestone[
        "country"] == "USA"].groupby([
    "city"])[
    "price"].max().sort_values(ascending=False)
```

```python
# Visualisasi data pada kota di USA dengan harga penjualan rumah terbesar hingga terkecil
data2.plot(
    kind="bar", figsize=(10, 5), title="Harga Maksimum rumah yang dijual di masing-masing kota di USA")
```

Dari data diketahui untuk kota yang memiliki harga rumah termahal adalah kota Kent di USA dengan harga senilai $26.590.000.

3. Data apakah yang paling memfaktori harga penjualan rumah?

```python
# Pembuatan kolom yang ingin dihitung korelasinya
columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
           'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement',
           'yr_built', 'yr_renovated']

# Pembuatan kolom untuk menyimpan hasil perhitungan korelasi
correlations = {}

# Proses perhitungan korelasi untuk masing-masing kolom
for column in columns:
    correlations[column] = data_milestone['price'].corr(data_milestone[column])
    
# Hasil perhitungan korelasi berdasarkan kolom yang ditentukan    
rows =[correlations['bedrooms'],correlations['bathrooms'],correlations['sqft_living'],correlations['sqft_lot'],
   correlations['floors'],correlations['waterfront'],correlations['view'],correlations['condition'],
   correlations['sqft_above'],correlations['sqft_basement'],correlations['yr_built'],correlations['yr_renovated']]
```

```python
# Visualisasi data untuk menampilkan perbandingan korelasi data harga dengan data kolom yang telah ditentukan
fig, ax = plt.subplots(figsize=(15,5))
colors = []
for value in rows:
    if value < 0:
        colors.append('red')
    else:
        colors.append('green')
ax.bar(columns, rows, color=colors)
ax.set_ylabel('Korelasi')
ax.set_title('Korelasi antara kolom dengan harga rumah')
plt.show()
```

Dari data didapat hasil yang paling memfaktori harga rumah adalah luas rumah itu sendiri.

5. Berapa rentan waktu rata-rata antara tahun pembangunan rumah dengan tahun renovasi rumah?

```python
# Membuat tabel dengan kondisi tahun renovasi rumah harus lebih besar dari nol
data_copy = data_milestone.copy()
data5 = data_copy[data_copy["yr_renovated"] > 0]
data55 = data5.drop(data5.columns[2:12], axis=1)
```

```python
# Proses pengkategorian data menjadi data numerik dan data kategorial
num_columns = data55.select_dtypes(include=np.number).columns.tolist()
cat_columns = data55.select_dtypes(include=['object']).columns.tolist()
# Menghitung nilai mean, median, dan mode
data_mean = data55[num_columns].mean().rename('mean')
data_median = data55[num_columns].median().rename('median')
data_mode = data55[num_columns + cat_columns].mode().iloc[0].rename('mode')
# Membuat tabel central tendency
central_tendency = pd.concat([data_mean, data_median, data_mode], axis=1)
central_tendency
```

```python
# Menghitung selisih rata-rata antara tahun pembangunan rumah dengan tahun renovasi rumah
selisih = round(data55['yr_renovated'].mean() - data55['yr_built'].mean())
print('Dari data didapat untuk waktu rata-rata antara tahun pembangunan dengan tahun renovasi rumah adalah sekitar',selisih,'tahun')
```

6. Kenapa ada rumah yang kondisi 1 (tidak layak untuk dijadikan tempat tinggal)?

Menurut saya karena nilai kondisi (view) sama dengan nol. Untuk menguji pendapat saya, saya akan menggunakan uji hipotesis dengan metode Single Sample Two Tailed.
Dimana untuk keterangan H0 dan H1 adalah sebagai berikut:
* H0: Rata-rata nilai pemandangan (view) yang diberikan = 0
* H1: Rata-rata nilai pemandangan (view) yang diberikan ≠ 0

```python
# Membuat tabel dengan nilai kondisi sama dengan satu
data6 = data_copy[data_copy["condition"] == 1]
```

```python
# Menghitung t-statistic dan p-value
t_stats, p_v = st.ttest_1samp(data6['view'], 0)
# Membuat kondisi dimana:
# jika p-value lebih kecil dari 0.05 maka H0 ditolak, 
# jika p-value lebih besar sama dengan 0.05 maka H0 gagal ditolak. 
if p_v < 0.05:
    print('Kita menolak H0. Tidak cukup bukti untuk menyatakan bahwa rata-rata nilai pemandangan (view) yang diberikan tidak sama dengan 0.')
else:
    print('Kita gagal menolak H0. Rata-rata nilai pemandangan (view) yang diberikan sama dengan 0.')
```

Dari data dapat ditarik kesimpulan bahwa Rata-rata nilai pemandangan (view) yang diberikan sama dengan 0 untuk kondisi rumah yang berada dalam penilaian 1, dengan kata lain pendapat saya benar.
